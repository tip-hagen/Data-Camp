**Guardrails & Safe Agents**

In this final chapter we explore how to build resilient, safe agentic systems. As agents gain power, guardrails become essential to prevent misuse and stay within design intent.


**Motivation & Examples**

Imagine an HR agent with access to employee records. It's fine answering "When is my next review?" but a query like "Where does my colleague live?" is a privacy violation. Or an unrelated request such as "Help me build a dashboard in Python" wastes resources. Without safeguards, even a helpful agent can become a liability.

Guardrails act like safety barriers, keeping the agent on track and protecting data and reputation.


**Types of Guardrails**

1. **Input Guardrails** – trigger on user prompts.
   - *Relevance classifier* redirects off-topic queries back to scope.
   - *Safety classifier* blocks prompt-injection attempts ("Forget instructions...").
   - *Moderation* filters hate speech or other harmful content.
   - *Rules-based* protections include blocklists, length limits, regex filters (e.g. >1000 words or forbidden terms).

2. **Tool-Based Guardrails** – activate when the model interacts with tools.
   - *Tool safeguards* assess risk levels (e.g. reading vacation balances vs altering salaries), and can pause high-risk actions for human approval.

3. **Output Guardrails** – check model responses before hitting users.
   - *PII filters* remove personally identifiable data (SSNs, addresses).
   - *Output validation* ensures tone and content align with brand/policy.


**Guardrails & the Agentic Trinity**

Where do guardrails fit in our architecture?

- Input guardrails screen requests **before** they reach the model.
- Tool guardrails monitor **during** tool invocations.
- Output guardrails filter responses **after** the model produces them.

The orchestration layer coordinates all guardrails, deciding when to block, modify, or escalate requests.


**Summary & Next Steps**

Guardrails are a core part of responsible agent design. Apply them during input processing, tool usage, and output generation to keep agents compliant, secure, and focused. As you build more advanced systems, revisit and extend these safeguards.

**Final Lesson: Agentic Systems in the Real World**

This lesson collects best practices for working with agentic systems, whether you're using off-the-shelf tools or building your own.

**Using Off-the-Shelf Tools**

- Provide rich context and well-designed prompts; include examples, style guidelines, and relevant information to guide the agent.
- Understand the agent's capabilities and limitations: know which tools it can call and how up-to-date its knowledge is.
- Always verify the agent's output—models can hallucinate or misinterpret complex data, so double-check important results.
- Be mindful of costs; many tools charge per-use or per-token, so balance performance with budget.
- Use agents responsibly; guard against leaking confidential or personally identifiable information to third parties.

**Designing & Building Agents**

- Design for human intervention: build clear escalation paths and mechanisms for the agent to signal uncertainty.
- Critically evaluate whether you need an agent. If the workflow is predictable and rule-based, traditional automation may suffice.
- Monitor costs and performance trade-offs; each reasoning cycle and tool call carries time and monetary expense.
- Start simple and iterate: begin with a single agent and basic tools, adding complexity only when needed.
- Monitor everything: track success rates, response times, costs, and user satisfaction to drive continuous improvement.

These guidelines round out your understanding and prepare you to apply agents effectively in real-world settings.
