**Foundations of AI Agents**

**What is AI Agents**

An **AI agent** is defined as a system capable of reasoning, planning, and acting to fulfill a user‑specified objective. The travel‑planning example “Jeeves” illustrates these capabilities: the agent

- understands natural-language requests,
- breaks tasks into steps and chooses appropriate tools, and
- interacts with external services (calendar, travel sites, etc.) to carry out the plan.

Formally, an agent uses an AI model to engage with its environment, combining its “brain” (the model responsible for reasoning and planning) with its “body” (external tools it can invoke). Agency exists on a spectrum—from basic chatbots with no real agency, through tool‑using agents like Jeeves, to advanced multi-agent systems coordinating many steps or actors.


**Key Concepts**

- **AI agent**: an AI system capable of reasoning, planning, and acting.
- **Brain vs. body**: model = reasoning/planning; tools = actions.
- **Agency spectrum**: ranges from non‑agentic systems (chatbots) to tool‑using agents, then multi‑step/multi‑agent systems.
- **Formal definition**: system using an AI model to interact with its environment for a user‑defined goal.
- **Example (Jeeves)**: Planning a trip as a concrete illustration of reasoning→planning→acting.



**Spectrum of Agency**

| Level | Agency | Description | Examples |
|-------|--------|-------------|----------|
| 0 | No agency | Systems that can only respond based on trained knowledge or perform discrete, pre-defined tasks | Chatbots with trained knowledge (e.g. GPT-4o), workflow automation systems |
| 1 | Basic routing | AI models that can route scenarios in workflows | A customer support workflow where an AI model routes a ticket based on its content |
| 2 | Tool-using agents | Systems that can utilize external tools | A travel AI agent that can book flights (e.g. Jeeves) |
| 3 | Autonomous agents | Systems that can perform multiple steps autonomously | Deep research tools that can perform multi-step reasoning and tool calling |
| 4 | Multi agent systems | Systems that can delegate workflows to multiple agents | Coding assistants that can ideate, generate, and push code to an existing codebase |

**Exercise: Know Your Agent**

You're consulting for a retail company that wants to augment their customer service function with AI. During the meeting, the CEO asks:

> "We keep hearing about AI agents, but what exactly are they? Our customer support team already uses AI to answer FAQs and workflow automation tools that process returns."

**Question:** What's the definition of an AI Agent?

- (1) An AI agent is an advanced chatbot that can answer any question using its trained knowledge base and provide more detailed responses than traditional FAQ systems.
- (2) An AI agent is an automation tool that can execute predefined workflows faster than humans.
- (3) **An AI agent is a system that leverages an AI model to interact with its environment to achieve a user-defined objective, combining reasoning, planning, and the execution of actions through external tools.**
- (4) An AI agent is any AI system that uses large language models to process text and generate human-like responses to user queries.

*Answer: (3)*


**Agentic Components**

AI agents are made agentic by three core components, often referred to as the Agentic Trinity:

- **Model (the brain)** – a large language model that interprets the request, reasons about it, and breaks the problem into smaller steps. In the travel example, the model figures out what needs to be done to book flights and hotels.
- **Tools (the body)** – external systems the model can call to interact with the real world. Jeeves, for instance, accesses calendars, company policy documents, and travel websites to gather information and perform actions.
- **Orchestration layer** – the control loop that manages the agent’s decision‑making cycle. It takes in new data, maintains context, decides the next action, and repeats until the objective is met or a stopping condition occurs. Orchestration can range from simple if–then rules to complex reasoning pipelines.

These elements work together continuously: the model reasons, tools execute the plan, and orchestration keeps the process moving.

**When to Use an Agentic System**

Not every problem requires an AI agent. The latest transcript explores scenarios where agents are appropriate versus simpler solutions.

- Agents excel when tasks involve **complex decision-making**, **unstructured data**, **hard-to-maintain rules**, or require **adaptive problem solving**.
- **Customer support team A** handles mostly predictable FAQ-style tickets; a simple chatbot suffices. 80 % of their issues are straightforward and do not require accessing user data or tools.
- **Customer support team B** faces intricate cases needing context, history, and multi-step remediation; here an agentic solution with tool access is beneficial.

**Team Comparison**

| Scenario | Agent Suitable? | Why |
|----------|----------------|-----|
| Team A (order/status questions) | No | Simple answers, no tool access needed |
| Team B (billing disputes, credits) | Yes | Requires data access, adaptive workflows |

**Tooling Spectrum**

The agent ecosystem spans several layers. Choose based on your needs:

```mermaid
flowchart LR
  OffShelf[Off-the-shelf] --> LowCode[Low-code/No-code]
  LowCode --> Frameworks[Agent frameworks]
```

- **Off-the-shelf**: ready-made agents for narrow domains (e.g., AI coding assistants).
- **Low-code/No-code**: customizable workflows with moderate complexity; accessible to business users.
- **Frameworks**: build from scratch when proprietary data, sensitivity, or full control are required.

**Build vs. Buy Guidelines**

- **Buy off-the-shelf** when a mature solution exists and you want low maintenance.
- **Buy low-code/no-code** when you need some customization without full engineering investment.
- **Build with frameworks** when your agent is core to your competitive advantage or integrates with sensitive systems.
