**AI Agents in the Wild**

*Course: Building Scalable Agentic Systems — Instructor: Korey Stegared-Pace, Senior AI Cloud Advocate at Microsoft*

**Overview**

An AI agent receives a user prompt, passes it to a Large Language Model (LLM) that interprets the task and determines which tools to invoke, then feeds tool results back into the LLM — a loop managed by an orchestration layer that handles memory and logging. A complete agentic application is more than just the agent loop: it includes a conversational interface, system-message prompt engineering, a remotely hosted model, and integrated data storage. Not every problem warrants an agent — agentic solutions are best suited to tasks that are open-ended, multi-step, and adaptive. The course also introduces interoperability frameworks — **Model Context Protocol (MCP)** by Anthropic and **Agent-to-Agent (A2A)** by Google — and production best practices for failure handling.

**Key Concepts**

- **AI Agent**: a system where an LLM interprets a task and autonomously decides which tools to call to complete it.
- **LLM (Large Language Model)**: the reasoning core of the agent — interprets prompts and generates plans or responses.
- **Tools**: functions that let the agent interact with the external world (query APIs, run code, fetch data, trigger events).
- **Orchestration Layer**: manages the agent loop, coordinates tool calls, and maintains memory and logging.
- **System Messages**: additional context injected alongside the user prompt to shape model behaviour.
- **Model Context Protocol (MCP)**: an interoperability framework by Anthropic for integrating agents with surrounding systems.
- **Agent-to-Agent (A2A)**: an interoperability framework by Google enabling agents to communicate with each other.
- **When to use agents**: best for problems that are *open-ended*, *complex/multi-step*, and *adaptable* — not for simple deterministic workflows.

**Design Principles for Scalable Agents**

*What makes agents scalable and the three core design principles for building them.*

**Overview**

Scalable agents are able to withstand the challenges of high user traffic while maintaining consistently high performance. High traffic introduces greater variety in user inputs, a higher likelihood of malicious actors, and significantly more model usage and tool calls. To mitigate these risks, three core design principles are proposed: **robust infrastructure and tooling**, **modular design architecture**, and **continuous evaluation and feedback loops**. Modularity — designing systems as separate, independent components — is illustrated through the LEGO analogy: changing one brick type does not require touching any other. For agents, this means both software components (UI, agents, data stores) and the agents themselves can be designed, maintained, and scaled independently.

**Key Concepts**

- **Scalable agent**: an agent that maintains consistently high performance under high user traffic, handling greater input variety, more malicious actors, and higher model/tool usage.
- **Robust infrastructure and tooling**: reliable compute, storage, deployment pipelines, and established agent tooling (tool calling, multi-step reasoning, memory management).
- **Compute**: computational resources for running agent code and workflow processes.
- **Storage**: stores agent states, user chat histories, and logs for monitoring and debugging.
- **Deployment pipelines**: mechanisms for applying updates and changes with low risk to user experience.
- **Modularity**: designing systems as separate, independent components that can be built and maintained without impacting each other.
- **Software modularity**: application components (UI, agents, data stores) designed as distinct units with their own logic — enabling easier updates and monitoring in isolation.
- **Multi-agent system**: multiple agents each operating over a clearly defined task domain, rather than one agent handling everything with many tools.
- **Continuous evaluation and feedback loops**: tracking success rates, latencies, and error patterns, plus collecting user feedback (e.g. thumbs-up/down) to improve prompts, tools, data, and parameters.

*Each pillar addresses a different dimension of scaling risk in agentic systems.*

|---|---|---|

**How AI Agents Scale (and Fail)**

*Common failure modes that emerge when scaling AI agents to real-world, high-traffic production environments.*

**Overview**

AI agents that perform well in development can become fragile at scale: real users send messy, multilingual, slang-filled inputs; they ask out-of-scope questions; feedback mechanisms can be gamed; and multi-step tool calls accumulate latency and cost. This lesson uses an AI travel assistant as a running example to illustrate five common failure modes — fragile evaluation, intent drift, undesirable feedback loops, latency bottlenecks, and cost explosion — and the architectural strategies used to address each one.

**Key Concepts**

- **Fragile evaluation**: testing only with ideal inputs; fix by using real, messy, multilingual queries and simulating global user interactions to expose hidden assumptions (time zones, currencies, cultural context).
- **Intent drift**: users asking the agent things outside its designed scope; fix by implementing **guardrails** that restrict the agent's operational boundaries and respond politely when out of scope.
- **Guardrails**: constraints that define and enforce the agent's scope of operation.
- **Undesirable feedback loops**: a rating system that optimises for engagement signals (e.g. jokes) over correctness; fix by blending human review with clearly defined metrics such as truthfulness, clarity, and tone.
- **Latency bottlenecks**: multi-step reasoning, tool calls, and retrieval accumulate delays at scale; fix by caching common queries, routing simple tasks to lighter models, and triggering heavy reasoning only when needed.
- **Cost explosion**: long prompts, multiple tools, advanced models, and external API calls compound rapidly with user volume; fix by applying cost-aware design early — asking whether steps can be cached, simplified, or skipped — and adding fair usage limits to costly features.

**Visual Aid: Five Failure Modes at Scale**

*Each failure mode, its cause in the travel assistant example, and the recommended fix.*

| # | Failure Mode | What Goes Wrong | Fix |
|---|---|---|---|
| 1 | **Fragile evaluation** | Real users send slang, fragments, emojis, non-English; agent makes hidden cultural assumptions | Test with real, messy, multilingual inputs; simulate global interactions |
| 2 | **Intent drift** | Users ask out-of-scope questions (lounges, restaurants, rails); agent guesses and gives bad answers | Implement guardrails to define scope; respond honestly when out of scope |
| 3 | **Undesirable feedback loops** | Rating system rewards jokes → agent prioritises charm over truth | Blend human review with metrics: truthfulness, clarity, tone |
| 4 | **Latency bottlenecks** | Multi-step reasoning + tool calls + retrieval feel slow at scale | Cache common queries; use lighter models for simple tasks; trigger heavy reasoning only when needed |
| 5 | **Cost explosion** | Long prompts, multiple tools, advanced models, external APIs multiply with user volume | Cost-aware architecture design early; add fair usage limits to costly features |

**Exercise: Don't Fail When You Scale!**

You're an AI Engineer at Nile, an online retailer and logistics company. The agentic support system has been exposed to 5% of platform users with great feedback and metrics. You're now planning to scale to 50% of users for an A/B test, and want to anticipate potential issues before doing so.

**Question:** Which of the following are common issues found when scaling? *(Select all correct answers)*

- (1) **Higher-than-expected costs** ✅
- (2) **Higher-than-expected latency** ✅
- (3) Improved multi-agent coordination ❌
- (4) **Increased error rate** ✅

*Answer: (1), (2), and (4) — cost explosion, latency bottlenecks, and increased error rates (from fragile evaluation and intent drift) are all documented failure modes at scale. Improved multi-agent coordination is a design benefit, not a scaling problem.*
