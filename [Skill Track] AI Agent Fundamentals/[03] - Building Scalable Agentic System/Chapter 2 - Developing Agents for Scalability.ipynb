**Multi-Agent Design Patterns**

*Practical strategies for developing agents for scale, focusing on network and supervisor multi-agent architectures.*

**Overview**

Single agents can handle many tasks when given multiple tools and clear prompts, but as the number of tools grows it becomes harder for the model to reason about which tool to use and in what order. Multi-agent systems solve this by assigning each agent a clearly scoped task domain with its own dedicated tools. Two leading patterns are covered: the **network (decentralised/swarm)** pattern, where agents peer-to-peer handoff tasks between themselves, and the **supervisor** pattern, where a central supervisor delegates to specialised worker agents and is solely responsible for responding to the user. Both patterns are illustrated with concrete examples — an employee onboarding system and a customer support system respectively — using terminology from LangChain.

**Key Concepts**

- **Single agent**: one agent with access to multiple tools; works well at small scale but struggles to reason over large tool sets in multi-step tasks.
- **Multi-agent system**: multiple specialised agents each operating over a clearly defined task domain with their own set of tools.
- **Network pattern (swarm / decentralised)**: user input is sent to a default agent; agents can autonomously handoff to one another to complete tasks and reach a solution — no central authority.
- **Handoff**: the act of one agent passing a task or query to another agent in a network architecture.
- **Supervisor pattern (hierarchical)**: a central supervisor agent receives user input, delegates tasks to specialised worker agents, collects their outputs, and is the only agent that responds to the user.
- **Worker agent**: a specialised agent in a supervisor architecture that executes tool calls and returns results to the supervisor.
- **Supervisor agent**: the orchestrating agent in a hierarchical architecture — it reasons about task requirements, delegates, checks completion, and generates the final user response.

**Visual Aid: Pattern Comparison**

*Key differences between the network and supervisor multi-agent architectures.*

| Dimension | Network (Swarm) | Supervisor (Hierarchical) |
|---|---|---|
| **Control** | Decentralised — agents decide handoffs autonomously | Centralised — supervisor controls all delegation |
| **User response** | Any agent can respond to the user | Only the supervisor responds to the user |
| **Agent communication** | Peer-to-peer handoffs | Workers report back to supervisor only |
| **Best for** | Open-ended exploration across domains | Structured tasks with clear delegation |
| **Example** | Employee onboarding (docs → messaging agent) | Customer support (supervisor + retrieval workers) |
